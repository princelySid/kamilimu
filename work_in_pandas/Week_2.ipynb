```python
bios_cleaned = bios_raw.copy()
```

```python
bios_cleaned['Roles'].str.split('')
```

```python
# actually unique rolls
bios_cleaned['Roles'].str.split('').explode().str.strip().unique()
```

```python
bios_cleaned['Roles'].str.split('').explode().str.strip().value_counts()
```

```python
bios_cleaned[['role1', 'role2', 'role3', 'role4', 'role5']] = bios_cleaned['Roles'].str.split('', expand=True)
bios_cleaned.head(2)
```

```python
# sex, is there more than 2?
bios_cleaned['Sex'].unique()
```

```python
# full names/used name first, name
bios_cleaned['Full name']
```

```python
bios_cleaned['Born']
```

```python
# born, split into date, place
bios_cleaned['Born']
date_pattern = r'(\d+ \w+ \d{4})'

bios_cleaned['date_born'] = bios_cleaned['Born'].str.extract(date_pattern)
bios_cleaned['date_died']= bios_cleaned['Died'].str.extract(date_pattern)
```

```python
#fix this for place born and died, doesn't necessarily need string match
country_pattern = r"(\(\w+]\))"
bios_cleaned['Born'].str.extract(country_pattern)
```

```python
# Split the Measurements column into height & weight columns
bios_cleaned[['height_cm', 'weight_kg']] = bios_cleaned['Measurements'].str.split('/', expand=True)
```

```python
bios_cleaned[['height_cm', 'weight_kg']]
```

```python
height = bios_cleaned[['NOC', 'height_cm']].copy()
height['height_cm']=height['height_cm'].str.extract(r'(\d+)')
height['height_cm']=pd.to_numeric(height['height_cm'], errors='coerce')

height.groupby('NOC')['height_cm'].mean().sort_values(ascending=False)
```
